# Singular Value Decomposition

generate 100x1000 matrix, get principle components

```python
import numpy as np
```

```python
vector1 = np.random.randint(-100, 101, size=1000)
vector2 = 2 * vector1
vector3 = np.random.randint(-100, 101, size=1000)
vector4 = -3 * vector3
vector5 = np.random.randint(-100, 101, size=1000)
vector6 = 1 * vector5
vector7 = np.random.randint(-100, 101, size=1000)
vector8 = 4 * vector7
vector9 = np.random.randint(-100, 101, size=1000)
vector10 = -2 * vector9

matrix_part1 = np.array([vector1, vector2, vector3, vector4, vector5, vector6, vector7, vector8, vector9, vector10])
matrix_part2 = np.random.randint(-20, 21, size=(45, 1000))
matrix_part3 = np.random.randint(-5, 6, size=(45, 1000))

A = np.vstack((matrix_part1, matrix_part2, matrix_part3)).T

U, s, V = np.linalg.svd(A, full_matrices=True)
s
```

# 데이터 생성 방법
### 100 차원 벡터 1000개로 이루어진 행렬 (1000 x 100)

#### 조건: 10개 (2개씩 짝지어 5쌍 일차식 관계), 45개 (-20, 20 범위 랜덤), 45개 (-5, 5 범위 랜덤)

1. 1000 차원의 벡터를 5개 생성한다(-100, 100 범위). 특정 상수를 각각 곱해서 5개 더 생성한다
2. 1000 차원의 벡터를 -20, 20 범위로 45개 생성한다.
3. 1000 차원의 벡터를 -5, 5 범위로 45개 생성한다.

모든 벡터를 더해서 행렬(100 x 1000)을 만든다. 그 행렬을 전치해서 행렬 A(1000 x 100)을 만든다.

```python
vector1 = np.random.randint(-100, 101, size=1000)
vector2 = 2 * vector1
vector3 = np.random.randint(-100, 101, size=1000)
vector4 = -3 * vector3
vector5 = np.random.randint(-100, 101, size=1000)
vector6 = 1 * vector5
vector7 = np.random.randint(-100, 101, size=1000)
vector8 = 4 * vector7
vector9 = np.random.randint(-100, 101, size=1000)
vector10 = -2 * vector9

matrix_part1 = np.array([vector1, vector2, vector3, vector4, vector5, vector6, vector7, vector8, vector9, vector10])
matrix_part2 = np.random.randint(-20, 21, size=(45, 1000))
matrix_part3 = np.random.randint(-5, 6, size=(45, 1000))

A = np.vstack((matrix_part1, matrix_part2, matrix_part3)).T

U, s, V = np.linalg.svd(A, full_matrices=True)
```

```python
def calculate_distance(X, r):
    X_hat = np.zeros(100)
    for i in range(r):
        X_hat = X_hat + np.dot(X, V[i]) * V[i]

    dist = np.linalg.norm(X - X_hat)
    return dist
```
